```python
import numpy as np
import pickle
from matplotlib import pyplot as plt
```

```python
LSROM_Err=pickle.load(open('LSROM_Err.p','rb'))
NMROM_Err=pickle.load(open('NMROM_Err.p','rb'))
```

```python
relErr_LSROM_g=1e-2*LSROM_Err['relErr_LSROM_g'][:4]
relErr_LSROM_lspg=1e-2*LSROM_Err['relErr_LSROM_lspg'][:4]
maxErr_LSROM_g=1e-2*LSROM_Err['maxErr_LSROM_g'][:4]
maxErr_LSROM_lspg=1e-2*LSROM_Err['maxErr_LSROM_lspg'][:4]
proErr_LSROM=1e-2*LSROM_Err['proErr_LSROM'][:4]
redDimArr_LSROM=LSROM_Err['redDimArr_LSROM'][:4]
```

```python
relErr_NMROM_g=1e-2*NMROM_Err['relErr_NMROM_g'][:4]
relErr_NMROM_lspg=1e-2*NMROM_Err['relErr_NMROM_lspg'][:4]
maxErr_NMROM_g=1e-2*NMROM_Err['maxErr_NMROM_g'][:4]
maxErr_NMROM_lspg=1e-2*NMROM_Err['maxErr_NMROM_lspg'][:4]
proErr_NMROM=1e-2*NMROM_Err['proErr_NMROM'][:4]
redDimArr_NMROM=NMROM_Err['redDimArr_NMROM'][:4]
```

```python
plt.figure()
plt.plot(redDimArr_LSROM,relErr_LSROM_g,redDimArr_LSROM,relErr_LSROM_lspg,redDimArr_LSROM,proErr_LSROM)
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error'])
plt.show()
```

```python
plt.figure()
plt.plot(redDimArr_NMROM,relErr_NMROM_g,redDimArr_NMROM,relErr_NMROM_lspg,redDimArr_NMROM,proErr_NMROM)
plt.legend(['NM-Galerkin','NM-LSPG','NM proj. error'])
plt.show()
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(1.5*6.4, 1.5*4.8))
plt.semilogy(redDimArr_LSROM,relErr_LSROM_g,'p',redDimArr_LSROM,relErr_LSROM_lspg,'*',redDimArr_LSROM,proErr_LSROM,'+--',linewidth=3,markersize=12)
plt.semilogy(redDimArr_NMROM,relErr_NMROM_g,'o',redDimArr_NMROM,relErr_NMROM_lspg,'v',redDimArr_NMROM,proErr_NMROM,'x--',linewidth=3,markersize=12)
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error','NM-Galerkin','NM-LSPG','NM proj. error'],fontsize='x-small')
plt.xlabel('Reduced dimension')
plt.ylabel('Avg. rel. error')
plt.show()
```

```python
plt.figure()
plt.plot(redDimArr_LSROM,maxErr_LSROM_g,redDimArr_LSROM,maxErr_LSROM_lspg,redDimArr_LSROM,proErr_LSROM)
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error'])
plt.show()
```

```python
plt.figure()
plt.plot(redDimArr_NMROM,maxErr_NMROM_g,redDimArr_NMROM,maxErr_NMROM_lspg,redDimArr_NMROM,proErr_NMROM)
plt.legend(['NM-Galerkin','NM-LSPG','NM proj. error'])
plt.show()
```
