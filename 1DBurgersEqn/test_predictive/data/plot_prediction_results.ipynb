```python
import numpy as np
from matplotlib import pyplot as plt
import time, sys
import pickle
from scipy.io import savemat, loadmat
```

```python
# load
resultDic=pickle.load(open('prediction_NM-ROM.p','rb'))
resultArrLSPG=resultDic['LSPG']
resultArrGalerkin=resultDic['Galerkin']
resultArr=pickle.load(open('prediction_NM-LSPG-HR.p','rb'))
ampArr = resultArr[:,0]

# savemat
savemat('Prediction1d.mat',{'resultArrNMLSPG':resultArrLSPG,'resultArrNMLSPGHR':resultArr})
```

```python
plt.rcParams.update({'font.size': 20})
fig = plt.figure(figsize=(1.5*6.4, 1.5*4.8))
ax = fig.gca()
right_side=ax.spines['right']
right_side.set_visible(False)
top_side=ax.spines['top']
top_side.set_visible(False)
plt.plot(ampArr,1e-2*resultArr[:,-2],'x--',ampArr,1e-2*resultArrLSPG[:,-2],'+--',linewidth=3,markersize=12)
plt.xlabel(r'$\mu$')
plt.ylabel('Avg. rel. error')
plt.legend(['NM-LSPG-HR','NM-LSPG'],frameon=False)
plt.show()

plt.rcParams.update({'font.size': 20})
fig = plt.figure(figsize=(1.5*6.4, 1.5*4.8))
ax = fig.gca()
right_side=ax.spines['right']
right_side.set_visible(False)
top_side=ax.spines['top']
top_side.set_visible(False)
plt.plot(ampArr,1e-2*resultArr[:,-1],'x--',ampArr,1e-2*resultArrLSPG[:,-1],'+--',linewidth=3,markersize=12)
plt.xlabel(r'$\mu$')
plt.ylabel('Max. rel. error')
plt.legend(['NM-LSPG-HR','NM-LSPG'],frameon=False)
plt.show()
```
