```python
import numpy as np
import pickle
from matplotlib import pyplot as plt
```

```python
LSROM_Err=pickle.load(open('LSROM_Err_nparam.p','rb'))
NMROM_Err=pickle.load(open('NMROM_Err_nparam.p','rb'))
```

```python
relErr_LSROM_g=LSROM_Err['relErr_LSROM_g']*1e-2
relErr_LSROM_lspg=LSROM_Err['relErr_LSROM_lspg']*1e-2
maxErr_LSROM_g=LSROM_Err['maxErr_LSROM_g']*1e-2
maxErr_LSROM_lspg=LSROM_Err['maxErr_LSROM_lspg']*1e-2
proErr_LSROM=LSROM_Err['proErr_LSROM']*1e-2
nparamArr_LSROM=LSROM_Err['nparamArr_LSROM']
```

```python
relErr_NMROM_g=NMROM_Err['relErr_NMROM_g']*1e-2
relErr_NMROM_lspg=NMROM_Err['relErr_NMROM_lspg']*1e-2
maxErr_NMROM_g=NMROM_Err['maxErr_NMROM_g']*1e-2
maxErr_NMROM_lspg=NMROM_Err['maxErr_NMROM_lspg']*1e-2
proErr_NMROM=NMROM_Err['proErr_NMROM']*1e-2
nparamArr_NMROM=NMROM_Err['nparamArr_NMROM']
```

```python
plt.figure()
plt.plot(nparamArr_LSROM,relErr_LSROM_g,nparamArr_LSROM,relErr_LSROM_lspg,nparamArr_LSROM,proErr_LSROM)
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error'])
plt.show()
```

```python
plt.figure()
plt.plot(nparamArr_NMROM,relErr_NMROM_g,nparamArr_NMROM,relErr_NMROM_lspg,nparamArr_NMROM,proErr_NMROM)
plt.legend(['NM-Galerkin','NM-LSPG','NM proj. error'])
plt.show()
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(1.5*6.4, 1.5*4.8))
plt.semilogy(nparamArr_LSROM,relErr_LSROM_g,'p',nparamArr_LSROM,relErr_LSROM_lspg,'*',nparamArr_LSROM,proErr_LSROM,'+--',linewidth=3,markersize=12)
plt.semilogy(nparamArr_NMROM,relErr_NMROM_g,'o',nparamArr_NMROM,relErr_NMROM_lspg,'v',nparamArr_NMROM,proErr_NMROM,'x--',linewidth=3,markersize=12)
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error','NM-Galerkin','NM-LSPG','NM proj. error'],fontsize='x-small')
plt.xlabel('Number of parameter instances')
plt.ylabel('Avg. rel. error')
plt.show()
```

```python
plt.figure()
plt.plot(nparamArr_LSROM,maxErr_LSROM_g,nparamArr_LSROM,maxErr_LSROM_lspg,nparamArr_LSROM,proErr_LSROM)
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error'])
plt.show()
```

```python
plt.figure()
plt.plot(nparamArr_NMROM,maxErr_NMROM_g,nparamArr_NMROM,maxErr_NMROM_lspg,nparamArr_NMROM,proErr_NMROM)
plt.legend(['NM-Galerkin','NM-LSPG','NM proj. error'])
plt.show()
```

```python
plt.rcParams.update({'font.size': 20})
fig = plt.figure(figsize=(1.5*6.4, 1.5*4.8))
ax = fig.gca()
right_side=ax.spines['right']
right_side.set_visible(False)
top_side=ax.spines['top']
top_side.set_visible(False)
plt.semilogy(nparamArr_LSROM,maxErr_LSROM_g,'p',nparamArr_LSROM,maxErr_LSROM_lspg,'*',nparamArr_LSROM,proErr_LSROM,'+--',linewidth=3,markersize=12)
plt.semilogy(nparamArr_NMROM,maxErr_NMROM_g,'o',nparamArr_NMROM,maxErr_NMROM_lspg,'v',nparamArr_NMROM,proErr_NMROM,'x--',linewidth=3,markersize=12)
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error','NM-Galerkin','NM-LSPG','NM proj. error'],frameon=False,fontsize='small')
plt.xlabel('Number of parameter instances')
plt.ylabel('Max. rel. error')
plt.show()
```
