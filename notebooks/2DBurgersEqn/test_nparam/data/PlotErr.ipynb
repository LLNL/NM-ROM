```python
import numpy as np
import pickle
from matplotlib import pyplot as plt
```

```python
LSROM_Err=pickle.load(open('LSROM_Err.p','rb'))
NMROM_Err=pickle.load(open('NMROM_Err.p','rb'))
```

```python
relErr_LSROM_g=1e-2*LSROM_Err['relErr_LSROM_g']
relErr_LSROM_lspg=1e-2*LSROM_Err['relErr_LSROM_lspg']
maxErr_LSROM_g=1e-2*LSROM_Err['maxErr_LSROM_g']
maxErr_LSROM_lspg=1e-2*LSROM_Err['maxErr_LSROM_lspg']
proErr_LSROM=1e-2*LSROM_Err['proErr_LSROM']
nparamArr_LSROM=LSROM_Err['nparamArr_LSROM']
```

```python
ind_LS=[0,1,2,3,4]
```

```python
relErr_NMROM_g=1e-2*NMROM_Err['relErr_NMROM_g']
relErr_NMROM_lspg=1e-2*NMROM_Err['relErr_NMROM_lspg']
maxErr_NMROM_g=1e-2*NMROM_Err['maxErr_NMROM_g']
maxErr_NMROM_lspg=1e-2*NMROM_Err['maxErr_NMROM_lspg']
proErr_NMROM=1e-2*NMROM_Err['proErr_NMROM']
nparamArr_NMROM=NMROM_Err['nparamArr_NMROM']
```

```python
from scipy.io import savemat, loadmat
savemat('LSROM_Err.mat',LSROM_Err)
savemat('NMROM_Err.mat',NMROM_Err)
```

# Avg. rel. error of u and v

```python
plt.figure()
plt.plot(nparamArr_LSROM[ind_LS],relErr_LSROM_g[0][ind_LS],nparamArr_LSROM[ind_LS],relErr_LSROM_lspg[0][ind_LS],nparamArr_LSROM,proErr_LSROM[0])
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error'])
plt.show()

plt.figure()
plt.plot(nparamArr_LSROM[ind_LS],relErr_LSROM_g[1][ind_LS],nparamArr_LSROM[ind_LS],relErr_LSROM_lspg[1][ind_LS],nparamArr_LSROM,proErr_LSROM[1])
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error'])
plt.show()
```

```python
plt.figure()
plt.plot(nparamArr_NMROM,relErr_NMROM_g[0],nparamArr_NMROM,relErr_NMROM_lspg[0],nparamArr_NMROM,proErr_NMROM[0])
plt.legend(['NM-Galerkin','NM-LSPG','NM proj. error'])
plt.show()

plt.figure()
plt.plot(nparamArr_NMROM,relErr_NMROM_g[1],nparamArr_NMROM,relErr_NMROM_lspg[1],nparamArr_NMROM,proErr_NMROM[1])
plt.legend(['NM-Galerkin','NM-LSPG','NM proj. error'])
plt.show()
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(1.5*6.4, 1.5*4.8))
plt.semilogy(nparamArr_LSROM[ind_LS],relErr_LSROM_g[0][ind_LS],'p',nparamArr_LSROM[ind_LS],relErr_LSROM_lspg[0][ind_LS],'*',nparamArr_LSROM,proErr_LSROM[0],'+--',linewidth=3,markersize=12)
plt.semilogy(nparamArr_NMROM,relErr_NMROM_g[0],'o',nparamArr_NMROM,relErr_NMROM_lspg[0],'v',nparamArr_NMROM,proErr_NMROM[0],'x--',linewidth=3,markersize=12)
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error','NM-Galerkin','NM-LSPG','NM proj. error'],fontsize='small')
plt.xlabel('Number of parameter instances')
plt.ylabel('Avg. rel. error of u')
plt.show()

plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(1.5*6.4, 1.5*4.8))
plt.semilogy(nparamArr_LSROM[ind_LS],relErr_LSROM_g[1][ind_LS],'p',nparamArr_LSROM[ind_LS],relErr_LSROM_lspg[1][ind_LS],'*',nparamArr_LSROM,proErr_LSROM[1],'+--',linewidth=3,markersize=12)
plt.semilogy(nparamArr_NMROM,relErr_NMROM_g[1],'o',nparamArr_NMROM,relErr_NMROM_lspg[1],'v',nparamArr_NMROM,proErr_NMROM[1],'x--',linewidth=3,markersize=12)
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error','NM-Galerkin','NM-LSPG','NM proj. error'],fontsize='small')
plt.xlabel('Number of parameter instances')
plt.ylabel('Avg. rel. error of v')
plt.show()
```
