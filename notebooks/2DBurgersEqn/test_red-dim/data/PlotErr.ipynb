# Avg. rel. error of u and v

```python
plt.figure()
plt.plot(redDimArr_LSROM[ind_LS],relErr_LSROM_g[0][ind_LS],redDimArr_LSROM[ind_LS],relErr_LSROM_lspg[0][ind_LS],redDimArr_LSROM,proErr_LSROM[0])
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error'])
plt.show()

plt.figure()
plt.plot(redDimArr_LSROM[ind_LS],relErr_LSROM_g[1][ind_LS],redDimArr_LSROM[ind_LS],relErr_LSROM_lspg[1][ind_LS],redDimArr_LSROM,proErr_LSROM[1])
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error'])
plt.show()
```

```python
plt.figure()
plt.plot(redDimArr_NMROM,relErr_NMROM_g[0],redDimArr_NMROM,relErr_NMROM_lspg[0],redDimArr_NMROM,proErr_NMROM[0])
plt.legend(['NM-Galerkin','NM-LSPG','NM proj. error'])
plt.show()

plt.figure()
plt.plot(redDimArr_NMROM,relErr_NMROM_g[1],redDimArr_NMROM,relErr_NMROM_lspg[1],redDimArr_NMROM,proErr_NMROM[1])
plt.legend(['NM-Galerkin','NM-LSPG','NM proj. error'])
plt.show()
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(1.5*6.4, 1.5*4.8))
plt.semilogy(redDimArr_LSROM[ind_LS],relErr_LSROM_g[0][ind_LS],'p',redDimArr_LSROM[ind_LS],relErr_LSROM_lspg[0][ind_LS],'*',redDimArr_LSROM,proErr_LSROM[0],'+--',linewidth=3,markersize=12)
plt.semilogy(redDimArr_NMROM,relErr_NMROM_g[0],'o',redDimArr_NMROM,relErr_NMROM_lspg[0],'v',redDimArr_NMROM,proErr_NMROM[0],'x--',linewidth=3,markersize=12)
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error','NM-Galerkin','NM-LSPG','NM proj. error'],fontsize='small')
plt.xlabel('Reduced dimension')
plt.ylabel('Avg. rel. error of u')
plt.xticks([5,10,15,20])
plt.show()

plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(1.5*6.4, 1.5*4.8))
plt.semilogy(redDimArr_LSROM[ind_LS],relErr_LSROM_g[1][ind_LS],'p',redDimArr_LSROM[ind_LS],relErr_LSROM_lspg[1][ind_LS],'*',redDimArr_LSROM,proErr_LSROM[1],'+--',linewidth=3,markersize=12)
plt.semilogy(redDimArr_NMROM,relErr_NMROM_g[1],'o',redDimArr_NMROM,relErr_NMROM_lspg[1],'v',redDimArr_NMROM,proErr_NMROM[1],'x--',linewidth=3,markersize=12)
plt.legend(['LS-Galerkin','LS-LSPG','LS proj. error','NM-Galerkin','NM-LSPG','NM proj. error'],fontsize='small')
plt.xlabel('Reduced dimension')
plt.ylabel('Avg. rel. error of v')
plt.xticks([5,10,15,20])
plt.show()
```
